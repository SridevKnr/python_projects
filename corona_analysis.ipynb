```python
total_confirmed = df.groupby("Country/Region")["Confirmed"].sum()
```

```python
#sorts the resulting Series in descending order based on the calculated sums.
```

```python
total_confirmed_asc = total_confirmed.sort_values(ascending = False).head(20).reset_index()
total_confirmed_asc
```

```python
x_axis = "Confirmed"
y_axis = "Country/Region"
plt.figure(figsize=(8, 6))
fig = sns.barplot(x = x_axis, y = y_axis, data = total_confirmed_asc)

fig.bar_label(fig.containers[0])

plt.xlabel("Total Confirmed Cases", fontsize=10)
plt.ylabel("Countries", fontsize=10)
plt.title("Top 20 Countries by Total Cases", fontsize=12)

 
```

```python
total_death = df.groupby("Country/Region")["Deaths"].sum()

total_deaths_asc = total_death.sort_values(ascending = False).head(10).reset_index()
```

```python
total_deaths_asc
```

```python
x_axis = "Deaths"
y_axis = "Country/Region"

plt.figure(figsize=(10, 6))

fig = sns.barplot(x = x_axis, y = y_axis, data = total_deaths_asc, color = "Purple")
fig.bar_label(fig.containers[0])

plt.xlabel("Total Deaths", fontsize=10)
plt.ylabel("Countries", fontsize=10)
plt.title(" Countries by Death Rate", fontsize=12)
```

```python
total_new_cases = df.groupby("Country/Region")["New cases"].sum()
y_a = total_new_cases.sort_values(ascending = False).head(10).reset_index()
total_newdeaths = df.groupby("Country/Region")["New deaths"].sum()
y_b = total_newdeaths.sort_values(ascending = False).head(10).reset_index()
```

```python
merged_data = pd.merge(y_a, y_b, on = "Country/Region", suffixes = ("_Cases", "_Deaths"))
plt.figure(figsize = (12, 6))
sns.set(style = "whitegrid")


ax = sns.barplot(data = merged_data, x = "Country/Region", y = "New deaths", color ="red" , label = "New deaths")
sns.barplot(data = merged_data, x = "Country/Region", y  ="New cases" , color = "skyblue", bottom = merged_data["New deaths"], label = "New Deaths")

# Add values inside the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

# Customize the plot
plt.xlabel("Country")
plt.ylabel("Total")
plt.title("Top 10 Countries with New Cases and New Deaths as stacked bar chart")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()

# Show the plot
plt.show()



```

```python
df.groupby("WHO Region")["Confirmed"].sum().sort_values(ascending = False).reset_index()
```

```python
reg = df.groupby("WHO Region")[["Confirmed", "Deaths"]].sum().sort_values(by=["Confirmed", "Deaths"], ascending=False).reset_index()
reg
```

```python

# Plot a donut chart
fig, ax = plt.subplots()
ax.pie(reg["Confirmed"], labels=reg["WHO Region"], autopct='%1.1f%%', startangle=90, pctdistance=0.85)
# Draw a circle to make it a donut chart
center_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(center_circle)

# Add a circle at the center for the "Deaths" data
ax.pie(reg["Deaths"], radius=0.7, colors=['white'], startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')


plt.title("WHO Region - Confirmed Cases and Deaths")
plt.show()


```

```python
reg = df.groupby("WHO Region")[["Confirmed", "Deaths"]].sum().sort_values(by=["Confirmed", "Deaths"], ascending=False)

# Create a grouped bar chart

ax = reg.plot(kind="bar", stacked=False)

plt.title("WHO Region - Confirmed Cases and Deaths")
plt.xlabel("WHO Region")
plt.ylabel("Count")
plt.xticks(rotation=45)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize = 8, weight = "bold")

plt.show()
```
